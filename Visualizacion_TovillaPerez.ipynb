# The information gather is related to the price stock of the three principal banks of USA and Mexico.

# It is obtained by the same sources in datasets by bank.

# For each dataset the 'Date' column is homologated to have the same data type.

## J. P. Morgan

```python
url_jpm = "https://raw.githubusercontent.com/danieltp92/Sources/main/JPM_Historical_Data.csv"
data_jpm = pd.read_csv(url_jpm)
data_jpm
```

```python
data_jpm.info()
```

```python
jpm_df = data_jpm.copy()
jpm_df['Date_time'] = jpm_df['Date'].astype('datetime64[ns]')
jpm_df.info()
```

```python
columns_name = ["Date_time", "Stock_Price_JPM(USD)", "Open_Stock_Price_JPM(USD)", "High_Stock_Price_JPM(USD)", "Low_Stock_Price_JPM(USD)", "Vol_Stock_JPM", "Change_Stock_JPM_Perc"]
jpm_df = jpm_df.reindex(columns=columns_name)
jpm_df
```

# The information is joined in one final dataset which is going to be used as the master data

```python
df_banks = pd.merge(jpm_df, bac_df, how='outer', on='Date_time')
df_banks = pd.merge(df_banks, wfc_df, how='outer', on='Date_time')
df_banks = pd.merge(df_banks, bbva_df, how='outer', on='Date_time')
df_banks = pd.merge(df_banks, sant_df, how='outer', on='Date_time')
df_banks = pd.merge(df_banks, bano_df, how='outer', on='Date_time')
df_banks
```

```python
df_banks.info()
```

# Once the data is in one dataset, a comparison between banks in USA is made to watch their stock behavior during the time.

```python
plt.plot(df_banks['Date_time'], df_banks['Stock_Price_JPM(USD)'], label='J. P. Morgan')
plt.plot(df_banks['Date_time'], df_banks['Stock_Price_BAC(USD)'], label='Bank of America')
plt.plot(df_banks['Date_time'], df_banks['Stock_Price_WFC(USD)'], label='Wells Fargo')
plt.legend(loc='upper left')
```

# The behavior of Wells Fargo is better than the other two, although Bank of America has been increasing their stock value during the pass of the time, having lower value lost.

```python
df_banks.plot(x="Date_time", y=["Open_Stock_Price_JPM(USD)", "High_Stock_Price_JPM(USD)", "Low_Stock_Price_JPM(USD)"], kind="bar")
plt.legend(loc='lower right')
plt.show()
```

# The compare between the Stock prices during the time, while the market open, when hits ther higher and lower value and the final stock value is difficult to track for the volume of the data.

```python
plt.scatter(df_banks['Date_time'], df_banks['Open_Stock_Price_JPM(USD)'], label='J.P.M. Open Stock')
plt.scatter(df_banks['Date_time'], df_banks['Stock_Price_JPM(USD)'], label='J.P.M. Stock')
plt.legend(loc='upper right')
plt.show()
```

# In the scatter plot you can observe that the value when the Market is open and the final value of the day, has not much variance, showing that the price is consistent during the day. But in the pass of the time, you can observe that the Stock value has some depresiation.

```python
sns.barplot(x=df_banks['Date_time'], y=df_banks['Low_Stock_Price_BBVA(MXN)'])
```
